## Task 1: Unit Converter

```python
def unit_converter():
    """
    Prompts the user to choose a conversion category and specific units,
    then converts the input value and displays the result.

    Handles:
    - Length: m to ft, ft to m
    - Weight: kg to lbs, lbs to kg
    - Volume: L to gal, gal to L

    Returns:
    float: The converted value, or None if an error occurs.
    """
    print("Select Category:\n1. Length\n2. Weight\n3. Volume")
    try:
        category = input("Enter category (1-3): ")

        if category == '1':
            print("1. Meters to Feet\n2. Feet to Meters")
            choice = input("Choice: ")
            val = float(input("Enter value: "))
            if choice == '1':
                result = val * 3.28084
                print(f"{val}m = {result:.2f}ft")
            else:
                result = val / 3.28084
                print(f"{val}ft = {result:.2f}m")

        elif category == '2':
            print("1. Kg to Lbs\n2. Lbs to Kg")
            choice = input("Choice: ")
            val = float(input("Enter value: "))
            if choice == '1':
                result = val * 2.20462
                print(f"{val}kg = {result:.2f}lbs")
            else:
                result = val / 2.20462
                print(f"{val}lbs = {result:.2f}kg")

        elif category == '3':
            print("1. Liters to Gallons\n2. Gallons to Liters")
            choice = input("Choice: ")
            val = float(input("Enter value: "))
            if choice == '1':
                result = val * 0.264172
                print(f"{val}L = {result:.2f}gal")
            else:
                result = val / 0.264172
                print(f"{val}gal = {result:.2f}L")
        else:
            print("Invalid category selected.")
            return None
        return result
    except ValueError:
        print("Error: Please enter numeric values only.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Run the task
unit_converter()
```

##Task 2: Mathematical Operations on a List

```python
def math_on_list():
    """
    Prompts user for a list of numbers and an operation (sum, average, max, min).
    Performs the operation and displays the result.
    """
    def get_sum(lst):
        """Returns the sum of elements in a list."""
        return sum(lst)

    def get_average(lst):
        """Returns the average of elements in a list."""
        return sum(lst) / len(lst) if lst else 0

    def get_max(lst):
        """Returns the maximum element in a list."""
        return max(lst)

    def get_min(lst):
        """Returns the minimum element in a list."""
        return min(lst)

    try:
        user_input = input("Enter numbers separated by spaces: ")
        nums = [float(x) for x in user_input.split()]

        if not nums:
            raise ValueError("The list is empty.")

        print("Select Operation:\n1. Sum\n2. Average\n3. Maximum\n4. Minimum")
        op = input("Choice (1-4): ")

        if op == '1':
            print(f"Sum: {get_sum(nums)}")
        elif op == '2':
            print(f"Average: {get_average(nums)}")
        elif op == '3':
            print(f"Maximum: {get_max(nums)}")
        elif op == '4':
            print(f"Minimum: {get_min(nums)}")
        else:
            print("Invalid operation.")
    except ValueError as e:
        print(f"Input Error: {e}")

# Run the task
math_on_list()
```

#4.2 Exercise on List Manipulation

```python
num_list = [1,2,3,4,5,6,7,8,9]

def extract_every_other(lst):
    return lst[::2]

def get_sublist(lst, start, end):
    return lst[start:end+1]

def reverse_list(lst):
    return lst[::-1]

def remove_first_last(lst):
    return lst[1:-1]

def get_first_n(lst, n):
    return lst[:n]

def get_last_n(lst, n):
    return lst[-n:]

def reverse_skip(lst):
    return lst[-2::-2]
```

##4.3 Exercise on Nested List

```python
lst =[[1, 2], [3, 4]]

def flatten(lst):
    return [item for sublist in lst for item in sublist]

def access_nested_element(lst, indices):
    val = lst
    for i in indices:
        val = val[i]
    return val

def sum_nested(lst):
    total = 0
    for i in lst:
        if isinstance(i, list):
            total += sum_nested(i)
        else:
            total += i
    return total

def deep_flatten(lst):
    flat = []
    for i in lst:
        if isinstance(i, list):
            flat.extend(deep_flatten(i))
        else:
            flat.append(i)
    return flat

def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat) / len(flat) if flat else 0
```
